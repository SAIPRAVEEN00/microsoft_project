```python
import pandas as pd 
import matplotlib.pyplot as plt 

import warnings 
warnings.filterwarnings('ignore')

microsoft_df=pd.read_csv('Microsoft_Data.csv')

y=microsoft_df['HasDetections']
```

```python
microsoft_df.drop(columns='HasDetections',inplace=True)
```

```python
missing_value_percentage_df=pd.DataFrame({'Column_Name':microsoft_df.columns,
                                          'Missing_value_count':microsoft_df.isnull().sum(),
                                          'Missing_value_percentage':microsoft_df.isnull().sum()/microsoft_df.shape[0]*100})


missing_value_percentage_df=missing_value_percentage_df.sort_values(by='Missing_value_percentage',ascending=False)


missing_value_percentage_df.head(10)
```

```python
cols_to_drop=[]

cols_with_na_grt_50=list(missing_value_percentage_df[missing_value_percentage_df['Missing_value_percentage']>50].index)

for col in microsoft_df.columns:
    if microsoft_df[col].nunique()==microsoft_df.shape[0] or microsoft_df[col].nunique()==1:
        cols_to_drop.append(col)
        
        
columns_to_drop=cols_to_drop+cols_with_na_grt_50
```

```python
microsoft_df.drop(columns=columns_to_drop,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(microsoft_df,y,test_size=0.2,random_state=42)

X_train
```

```python
cat_cols=[col for col in microsoft_df.columns if microsoft_df[col].dtype == 'object']
num_cols=[col for col in microsoft_df.columns if microsoft_df[col].dtype == 'int64' or 
                                                    microsoft_df[col].dtype == 'float64']
```

```python
for col in num_cols:
    X_train[col].fillna(X_train[col].mean(),inplace=True)
    X_test[col].fillna(X_train[col].mean(),inplace=True)
```

```python
for col in cat_cols:
    X_train[col].fillna(X_train[col].mode()[0],inplace=True)
    X_test[col].fillna(X_train[col].mode()[0],inplace=True)
```

```python
X_train['OrganizationIdentifier']=X_train['OrganizationIdentifier'].fillna(X_train['OrganizationIdentifier'].mean())
X_test['OrganizationIdentifier']=X_test['OrganizationIdentifier'].fillna(X_train['OrganizationIdentifier'].mean())
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
import numpy as np
scaler=StandardScaler()
minmaxscaler=MinMaxScaler()

for col in num_cols:
    X_train[col]=minmaxscaler.fit_transform(np.array(X_train[col]).reshape(-1,1))
    X_test[col]=minmaxscaler.transform(np.array(X_test[col]).reshape(-1,1))
```

```python
oe_train=pd.get_dummies(X_train[cat_cols])
oe_test=pd.get_dummies(X_test[cat_cols])
```

```python
X_train_oe,X_test_oe=oe_train.align(oe_test,join='inner',axis=1,fill_value=0)

X_train_final=pd.concat([X_train_oe,X_train[num_cols]],axis=1)

X_test_final=pd.concat([X_test_oe,X_test[num_cols]],axis=1)

X_test_final.shape
```

```python
from sklearn.linear_model import LogisticRegression

logreg=LogisticRegression()
logreg.fit(X_train_final,y_train)
```

```python
logreg_pred=logreg.predict(X_test_final)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
confusion_matrix(y_test,logreg_pred)

precision_score(y_test,logreg_pred)
recall_score(y_test,logreg_pred)
f1_score(y_test,logreg_pred)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier(criterion="entropy",splitter="best",random_state=10)

dtc.fit(X_train_final,y_train)

dtc_pred=dtc.predict(X_test_final)

dtc_pred

r2_score(y_test,dtc_pred)
```

```python
mean_squared_error(y_test,dtc_pred)
```
